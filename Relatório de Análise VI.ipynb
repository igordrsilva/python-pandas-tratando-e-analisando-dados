## Criando Novas Variáveis

Agora, vamos criar 3 novas variáveis para ajudar o cliente final na tomada de decisão. Estas, devem ser: Valor Bruto, Valor do m² e valor bruto do m².

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
dados
```

Vamos criar uma nova variável para saber qual o valor bruto que deve ser pago para cada imóvel.

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

Acabamos de criar uma nova coluna que consta a soma do Valor do Aluguel + Condomínio + IPTU. 

```python
dados
```

Também podemos criar uma coluna para visualizar o valor do metro quadrado por imóvel.

```python
dados['Valor do m²'] = dados['Valor'] / dados['Area']
```

```python
dados
```

```python
# Como a visualização do número ficou com muitas casas decimais, podemos usar o .round(2) para arrendndar para 2 casa deciamis.
dados['Valor do m²'] = dados['Valor do m²'].round(2)
```

```python
dados
```

Acima, criamos o valor do m² a partir do valor do aluguel, mas tmb podemos criar a partir do valor bruto.

```python
dados['Valor m² Bruto'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados
```

```python
# Criamos uma lista com os tipos de casa, para posteriormente poder classificar se o item está nesta lista.
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# Criamos uma nova coluna para filtrar em Casas e Apartamentos
# para isso, criamos uma função lambda, onde há uma variável x, que é o item.
# Assim, se o valor na coluna Tipo da ocorrência x estiver na lista 'Casa', então, nessa nova coluna, esta ocorrência
# terá valor 'Casa', se não, terá valor 'Apartamento'
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
# Filtramos e usamos o .apply() para aplicar uma condição
```

```python
dados.head(10)
```

## Excluindo Variáveis

Podemos observar alguns casos onde não há dados de IPTU para alguns imóveis e isso compromete a validade dos nossos dados de Valor Bruto e Valor m² Bruto, Por isso, vamos excluir esses dados do nosso banco de dados.

```python
# Criamos um DF auxiliar apenas com as colunas adicionadas recentemente.
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor do m²', 'Valor Bruto', 'Valor m² Bruto']])
```

```python
dados_aux
```

```python
# Uma forma fácil de excluir uma variável do nosso DF é usar o del e passar o nome da variável no nosso DF
del dados_aux['Valor Bruto']
```

```python
dados_aux
```

```python
dados_aux.pop('Valor m² Bruto')
# Aqui usando o .pop() e passando o nome da variável. ele nos retorna uma Series com os itens que foram deletados do DF
```

```python
dados_aux
```

```python
dados.drop(['Valor Bruto', 'Valor m² Bruto'], axis = 1, inplace = True)
```

```python
dados
```

```python
dados.to_csv('aluguel_residencial.csv', sep = ';', index = False)
```
